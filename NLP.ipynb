```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
with open("Article.txt","r") as doc:
    docum=doc.read()
```

```python
print(docum)
```

```python
doc = nlp(docum)
```

Converting the input text to a list of sentences and computing the number of sentences in the given Text

```python
sentences = list(doc.sents)
sentences
```

```python
print('number of sentences is:',len(sentences),'\n')
```

```python
from collections import Counter
words = [token.text for token in doc
          if not token.is_stop and not token.is_punct]

word_freq = Counter(words)
word_freq 
```

Calculate the frequency of words in each sentence: the output should be a dictionary where each key is a sentence and the value is also a dictionary of word frequency.

```python
words=[]
for i in range(0,len(sentences)):
    words.append([token.text for token in sentences[i]])
        
          
words
```

```python
diction={}
for i in range(len(sentences)):
    diction[sentences[i]]=Counter(words[i])
    
diction    
```

Calculate Term frequency for each word in a sentence:

```python
tot=0
for k,v in dict.items(diction):
    print(k, v)    
    for p,s in dict.items(v):
        tot=tot+s
    print("\n Total Word in this sentence is ",tot)
    for w,r in dict.items(v):    
        print('TF of ',w,'is ')
        print(r/tot)
       
    tot=0   
```
